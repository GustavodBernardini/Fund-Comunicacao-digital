###1) Modelagem de uma variável aleatória: caso de um dado justo.
####1.1)

```python
N=10 #Vezes que o dado é jogado
amostras_dado6lados = np.random.uniform(0, 6, N) #distribuição uniforme de [0,6]
histogram_dado=np.histogram(amostras_dado6lados, bins=[0, 1, 2, 3, 4, 5, 6])
espacoAmostral_dado=np.arange(1, 7, 1) #vetor de 1 a 6, passo de 1
plt.stem(espacoAmostral_dado, histogram_dado[0], linefmt='b-', markerfmt='bo', basefmt='k-')
plt.xlabel("Faces do dado", fontsize=14)
plt.ylabel("Quantidade de ocorrência", fontsize=14)
plt.xlim(0, 7)
title = "Item 1.1) " + str(N) + " realizações da variável aleatória"
plt.title(title)
plt.show()
```

####1.2) 

```python
N=10 #Vezes que o dado é jogado
amostras_dado6lados = np.random.uniform(0, 6, N) #distribuição uniforme de [0,6]
#histogram agora é configurado com normed=1 para já normalizar o resultado
histogram_dado=np.histogram(amostras_dado6lados, bins=[0, 1, 2, 3, 4, 5, 6], normed=1) 
espacoAmostral_dado=np.arange(1, 7, 1) #vetor de 1 a 6, passo de 1
plt.stem(espacoAmostral_dado, histogram_dado[0], markerfmt='o', label='Estimada')
plt.stem(espacoAmostral_dado, [1/6,1/6,1/6,1/6,1/6,1/6], markerfmt='o', label='Teórica')
plt.legend()
plt.xlabel("Faces do dado", fontsize=14)
plt.ylabel("Quantidade de ocorrência normalizada", fontsize=10)
plt.xlim(0, 7)
plt.ylim(0,1) #acrescentado um limite para o eixo y de 1 já que 0 < probabilidade < 1
title = "Item 1.2) Estimação da pdf com " + str(N) + " realizações da variável aleatória" \
+ " e pdf teórica"
plt.title(title)
plt.show()
```

As pdfs tem um consideravel diferença devido ao número de amostras ser muito baixo para o caso simulado. Quanto maior a quantidade de realizações, mais a pdf estimada e teórica serão semelhantes.

1.3) Repetindo o processo de 1.2) nota-se grande variação na pdf estimada em cada caso, mas a medida que é selecionado um N muito grande, cada conjunto de N realizações acaba tendo praticamente o mesmo resultado na estimação da pdf da variável aleatória.

###2) Geração de uma variável aleatória: jogar um dado desbalanceado.

Se P(6)=1/4=0.25, P(1)+P(2)+P(3)+P(4)+P(5)=0.75
P(i| i<=5)=0.75/5=0.15

```python
N=10 #Vezes que o dado é jogado
amostras_dado6lados_desbalaceado = np.random.uniform(0, 100, N) #distribuição uniforme de [0,100]
histogram_dado=np.histogram(amostras_dado6lados_desbalaceado, bins=[0, 15, 30, 45, 60, 75, 100])
espacoAmostral_dado=np.arange(1, 7, 1) #vetor de 1 a 6, passo de 1
plt.stem(espacoAmostral_dado, histogram_dado[0]/N, markerfmt='o', label='Estimada')
plt.stem(espacoAmostral_dado, [0.15,0.15,0.15,0.15,0.15,0.25], markerfmt='o', label='Teórica')
plt.legend()
plt.xlabel("Faces do dado", fontsize=14)
plt.ylabel("Quantidade de ocorrência normalizada", fontsize=10)
plt.xlim(0, 7)
plt.ylim(0,1)
title = "Questão 2) Estimação da pdf com " + str(N) + " realizações da variável aleatória" \
+ " e pdf teórica"
plt.title(title)
plt.show()
```

### 3) Modelagem de uma variável aleatória normalmente distribuída X com média zero e variância unitária.

```python
N=1000 #numero de amostras 
media_X=0
std_X=1
amostras_X = np.random.normal(media_X, std_X, N)
media_X_medida = np.mean(amostras_X)
std_X_medido = np.std(amostras_X)
variancia_X_medida = np.var(amostras_X)
print("Questão 3")
print("Média estimada: ", media_X_medida)
print("Desvio padrão estimado: ", std_X_medido)
print("Variância estimada: ", variancia_X_medida)
```

```python
bins_hist_questao3 = np.arange(-4*std_X + media_X, 4*std_X + media_X, 0.1) #praticamente todos valores para a VA
plt.hist(amostras_X, bins_hist_questao3, normed=1)
plt.xlabel("Valor das amostras", fontsize=14)
plt.ylabel("Quantidade de ocorrência normalizada", fontsize=10)
plt.xlim(-4, 4)
plt.ylim(0,1)
title = "Questão 3) Estimação da pdf com " + str(N) + " realizações da variável aleatória" \
+ " com distribuição normal"
plt.title(title, fontsize=9)
plt.show()
```

###6) Experimento para testar o Teorema do Limite Central.
